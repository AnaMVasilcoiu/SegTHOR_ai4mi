```python
import matplotlib.pyplot as plt
import numpy as np

lrs = ["Baseline", "Combination 1", "Combination 2"]

labels = []
values = []
for lr in lrs:
    label = lr
    labels.append(label)

    if lr == "Baseline":
        path = "results/baseline100epochs/dice_val.npy"
        with open("results/baseline100epochs/best_epoch.txt", 'r') as file:
            best_epoch = int(file.read().strip())
    elif lr == "Combination 1":
        path = "results/AdamW/combination1-bestlr,bestK,bestinitK/dice_val.npy"
        with open("results/AdamW/combination1-bestlr,bestK,bestinitK/best_epoch.txt", 'r') as file:
            best_epoch = int(file.read().strip())
    else:
        path = "results/AdamW/combination2-bestlrbestK/dice_val.npy"
        with open("results/AdamW/combination2-bestlrbestK/best_epoch.txt", 'r') as file:
            best_epoch = int(file.read().strip())


    dice_val = np.load(path)
    score = dice_val[best_epoch, :, :].mean()
    values.append(score)

# Creating the horizontal bar plot
plt.figure(figsize=(6, 2))
bars=plt.barh(labels, values, color='mediumslateblue')
plt.xlim(0.8, 0.89)  # Set x-axis range to display scores from 0.8 to 0.9
plt.title("Dice Score in Validation for 100 epochs")
# plt.ylabel("Optimizer")
plt.xlabel("Dice Score")
plt.gca().invert_yaxis()  # Optional: Invert y-axis for better readability
for bar, value in zip(bars, values):
    plt.text(value + 0.001, bar.get_y() + bar.get_height() / 2, f'{value:.4f}', 
             va='center', ha='left', color='black')
plt.show()
```

```python
lrs = [1, 2, 2.5, 3, 3.5, 7]

import matplotlib.pyplot as plt
import numpy as np

labels = []
values = []
for lr in lrs:
    label = "0.000" + str(lr)
    if lr==2.5 or lr==3.5:
        label = "0.000" + str(lr)[0] + str(lr)[2]
    elif lr ==1:
        label = "0.0001 (baseline)"
    labels.append(label)

    if lr == 1:
        path= "results/baseline100epochs/"+ "/dice_val.npy"
        with open("results/baseline100epochs/" +'/best_epoch.txt', 'r') as file:
            best_epoch = int(file.read().strip())
    else:
        path = "results/AdamW/lr" + str(lr) + "/dice_val.npy"
        with open("results/AdamW/lr" + str(lr) +'/best_epoch.txt', 'r') as file:
            best_epoch = int(file.read().strip())
    
    dice_val = np.load(path)
    score =  dice_val[best_epoch, :, :].mean()
    values.append(score)

# Creating the vertical bar plot
plt.figure(figsize=(8, 6))
bars = plt.bar(labels, values, color='skyblue')
plt.ylim(0.82, 0.91)  # Set y-axis range from 0 to 1
plt.title("Dice score in Validation for 100 epochs with AdamW")
plt.xlabel("Learning rate")
plt.ylabel("Dice Score")
for bar, value in zip(bars, values):
    plt.text(bar.get_x() + (bar.get_width() / 2), value + 0.002,  f'{value:.4f}', 
             va='center', ha='center', color='black')
plt.show()
```
